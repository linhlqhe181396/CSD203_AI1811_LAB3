```python
#q1
def sum(n):
    if n == 1:
        return 1
    else:
        return n + sum(n - 1)
#q2    
def find_min(arr, n):
    if n == 1:
        return arr[0]
    min_subarray = find_min(arr, n - 1)
    return min(min_subarray, arr[n - 1])
#q3
def find_sum(arr, n):
    if n == 1:
        return arr[0]
    sum_subarray = find_sum(arr, n - 1)
    return sum_subarray + arr[n - 1]
#q4
def is_palindrome(arr, start, end):
    if start >= end:
        return 1
    if arr[start] != arr[end]:
        return 0
    return is_palindrome(arr, start + 1, end - 1)
#q5
def binary_search(arr, target, left, right):
    if left > right:
        return -1  
    mid = (left + right) // 2  
    if arr[mid] == target:
        return mid 
    elif arr[mid] > target:
        return binary_search(arr, target, left, mid - 1) 
    else:
        return binary_search(arr, target, mid + 1, right)
#q6    
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)
#q7
def power(x, n):
    if n == 0:
        return 1
    elif n < 0:
        return 1 / power(x, -n)
    else:
        return x * power(x, n - 1)
#q9
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
#q10
def addReciprocals(n):
    if n == 1:
        return 1.0
    else:
        return 1.0 / n + addReciprocals(n - 1)

```
